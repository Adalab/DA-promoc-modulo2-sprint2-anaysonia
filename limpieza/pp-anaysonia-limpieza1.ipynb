2. ¿Hay valores duplicados en nuestro dataframe? En caso de que los haya, eliminándlos.

```python
#Obtenemos las filas que estan dupicadas en nuestro dataframe.
df_preguntas[df_preguntas.duplicated()==True]
```

```python
#Observamos cuantas filas tenemos antes de eliminar los duplicados
print(f'Nuestro DataFrame antes de eliminar duplicados tiene {df_preguntas.shape[0]} filas y {df_preguntas.shape[1]} columnas')
```

```python
df_preguntas.drop_duplicates(inplace=True)
df_preguntas.head()
```

```python
#Observamos cuantas filas tenemos despues de eliminar los duplicados.
print(f'Nuestro DataFrame antes de eliminar duplicados tiene {df_preguntas.shape[0]} filas y {df_preguntas.shape[1]} columnas')
```

3. Como hemos visto, algunas columnas no tienen el tipo de datos que deberían. Cambiad el tipo de dato para la columna de year.

```python
df_preguntas['year'].dtypes
```

```python
df_preguntas['year']=df_preguntas['year'].astype('int64', errors='ignore')
df_preguntas.head()
```

```python
df_preguntas['year'].dtypes
```

4. En la columna de country poner todos los valores en minúsculas.
💡 Pista Tendréis que usar una función o una lambda 💡

```python
def pais_minusculas(columna):
    return str(columna).lower()
```

```python
df_preguntas['country']=(df_preguntas['country'].apply(pais_minusculas))
df_preguntas.head()
```
